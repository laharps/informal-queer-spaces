# Natural Language Processing

Next, we are going to find queer vocabulary from reviews of the businesses we already identified as queer. Additionally, we'll test our own list of queer vocabulary against these reviews. Ultimately, we will create a list of queer vocabulary to apply to the entire Philadelphia dataset.


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

import re
nltk.download('stopwords')
nltk.download('punkt')
```

```python
reviews = pd.read_pickle('data/gay_reviews_phil.pandas')
reviews.reset_index(inplace=True)
```

Let's look at the reviews dataframe.

```python
print('There are '+str(len(reviews)) + ' total reviews for '+ str(len(reviews.business_id.unique())) + ' unique businesses.')
print(reviews.text.head())
```

First, I'll put all the reviews into one massive string.

```python
s = ""
for i in range(len(reviews)):
    s+=reviews['text'][i]
```

Now split up all the words, clean out symbols and stopwords, and count them!

```python
def countWords(wordlist):
    counts = {} 
    
    for word in wordlist:
        lword = word.lower()
        if lword in counts:
            counts[lword] +=1
        else:
            counts[lword] = 1

    df = pd.DataFrame.from_dict(counts, orient='index', columns=['word_count'])
    df.sort_values('word_count', ascending=False, inplace=True)
    df.index.name = 'word'
    
    return df

wordlist = word_tokenize(re.sub(r"[^A-z\s]", "", s))

df = countWords(wordlist)
df.drop(index=stopwords.words('english'), errors='ignore', inplace=True)

df.head(15)
```

Unsurpisingly, we don't see any explicit gay words among the highest word count. Instead, we'll generate our own list of queer vocabulary to check for those.

```python
gay_vocab = ['gay', 'gays', 'queer', 'queers', 'lesbian', 'lesbians', 'dyke', 'dykes','lgbt', 'lgbtq', 
             'homosexual', 'homosexuals', 'homo', 'homos', 'homophobic', 'drag', 
             'queen', 'queens', 'trans', 'transgender', 'transphobic', 'bisexual','bisexuals',
             'twink', 'twinks', 'bear', 'bears']
```

```python
gay_word_count = df[df.index.isin([word for word in gay_vocab if word in df.index])] 
gay_word_count
```

# Run on All Businesses in Philadelphia
Using the list of queer words generated previously, we will identify other businesses that have those words in their reviews and count how often they appear in each one. Using the results, we will generate a dataframe of our new gay businesses.

Again, we won't run this step because the review dataframe is huge. We'll read in the relevant pickles at the end.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import ngrams, FreqDist

import re
nltk.download('stopwords')
nltk.download('punkt')

from collections import Counter
```

```python
# bring in all reviews in philadelphia
reviews = pd.read_pickle('data/reviewsPhil.pandas')

# standardize the reviews text
reviews['text'] = reviews['text'].fillna('').astype(str)

# group by business_id and sum
bizDf = reviews.groupby('business_id')['text'].sum().to_frame()
```

```python
# Create function to clean and count

def FindNewGays(in_df):
    
    # clean all the text and tokenize
    df = in_df.text.apply(lambda item: item.lower()).apply(lambda item: word_tokenize(re.sub(r"[^A-z\s]", "", item))).to_frame()
    
    # remove stopwords
    df['text'] = df['text'].apply(lambda review: [word for word in review if (word not in mystop)])

    # make column with gay words extracted
    df['gay_text'] = df['text'].apply(lambda review: [word for word in review if (word in gay_words)])

    # create new dataframe only with businesses that have at least one gay word in the reviews
    newgayDf = df[df['gay_text'].apply(lambda x: len(x) >0)].copy()

    # count number of words for each business and sort
    newgayDf['gay_text_len'] = newgayDf.gay_text.apply(lambda x: len(x))
    newgayDf.sort_values('gay_text_len', ascending=False, inplace=True)

    # create frequency distribution of gay words for each business
    newgayDf['gay_counts'] = newgayDf['gay_text'].apply(lambda words: Counter(words))

    # count unique gay words for each business
    newgayDf['gay_unique_count'] = newgayDf['gay_counts'].apply(lambda freq: len(freq.keys()))

    return newgayDf
```

```python
# Run function and create newgaybiz.pandas

# remove queen, bear, bears
gay_words = ['gay','drag','queer','queens','lgbt','lesbian','gays','lgbtq','homophobic','trans','lesbians','transgender','dyke','twinks','queers','homosexual','bisexual','homosexuals']
mystop = stopwords.words('english')

newgayDf = FindNewGays(bizDf)

newgayDf.to_pickle('data/newgaybiz.pandas')
```

# Analysis of New Gay Businesses
In the previous step, we have a list of business ids and some information about the gay words in the review. However, we don't have any other information about the businesses. We'll join this dataframe back to the original business dataframe to bring in additional info for analysis.

```python
# Bring in new gay businesses from last step
gaybizDf = pd.read_pickle('data/newgaybiz_clean.pandas')
gaybizDf.info()
```

```python
# Bring in all businesses in Philadelphia
phillybizDf = pd.read_pickle('data/bizPhil.pandas')
phillybizDf.info()
```

```python
# Join the two dataframes together
joinedDf = pd.merge(gaybizDf, phillybizDf, how='inner', on='business_id')
joinedDf.head()
```

We're interested in the category of businesses we have in our new dataset as compared to our original dataset of known gay businesses. Unfortunately, the dataframe does not have a simple category, but rather a list of various tags. We'll do a frequency analysis of the tags to figure out the most common business types.

```python
# Drop NAs and make a list of the tags
queercats = joinedDf.categories.dropna()

# Put all the tags in a single string
s = ""
for i in range(len(queercats)):
    s+=queercats[i]
    
s[0:1000]
```

```python
# Run frequency analysis of tags
def countWords(wordlist):
    counts = {} 
    
    for word in wordlist:
        lword = word.lower()
        if lword in counts:
            counts[lword] +=1
        else:
            counts[lword] = 1

    df = pd.DataFrame.from_dict(counts, orient='index', columns=['word_count'])
    df.sort_values('word_count', ascending=False, inplace=True)
    df.index.name = 'word'
    
    return df

wordlist = word_tokenize(re.sub(r"[^A-z\s]", "", s))

df = countWords(wordlist)
df.drop(index=stopwords.words('english'), errors='ignore', inplace=True)

df.head(10)
```

Let's visualize the top categories with a histogram.

```python
# Create histogram of top tags
import seaborn as sns

fig, ax = plt.subplots(figsize=(4,10))
sns.barplot(x=df.word_count.values[:20], y=df.index[:20], ax=ax)
ax.set_title("Number of Business with Tag")
ax.set_ylabel("Tag")
```

Looks like bars and restaaurants are the top categories, however we also see services, event, planning, and venues. Now we'll compare this to a histogram of our original queer business dataset to see if we captured any new businesses. 

```python
#lets compare the tags from the originally visible queer spaces
gay_orig = joinedDf[joinedDf.known==True]
gay_orig = gay_orig.categories.dropna()
gay_orig.reset_index(drop=True, inplace=True)

s = ""
for i in range(len(gay_orig)):
    s+=gay_orig[i]

print(s)
```

```python
def countWords(wordlist):
    counts = {} 
    
    for word in wordlist:
        lword = word.lower()
        if lword in counts:
            counts[lword] +=1
        else:
            counts[lword] = 1

    df2 = pd.DataFrame.from_dict(counts, orient='index', columns=['word_count'])
    df2.sort_values('word_count', ascending=False, inplace=True)
    df2.index.name = 'word'
    
    return df2

wordlist = word_tokenize(re.sub(r"[^A-z\s]", "", s))

df2 = countWords(wordlist)
df2.drop(index=stopwords.words('english'), errors='ignore', inplace=True)
```

```python
# plot
fig, ax = plt.subplots(figsize=(4,10))
sns.barplot(x=df2.word_count.values[:20], y=df2.index[:20], ax=ax)
ax.set_title("Number of Business with Tag")
ax.set_ylabel("Tag")
```

We see a lot more bars and nightlife in our initial dataset (clubs, dance, music). Our analysis worked! We found a wider variety of queer businesses than those that are formally tagged. Our method likely cast too wide of a net, so further work should be conducted to refine the queer vocabularly list and narrow down the dataset.
